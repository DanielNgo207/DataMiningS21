```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_2014 = pd.read_csv('./new_data/2014_US_exchange.csv', index_col=0)
df_2015 = pd.read_csv('./new_data/2015_US_exchange.csv', index_col=0)
df_2016 = pd.read_csv('./new_data/2016_US_exchange.csv', index_col=0)
df_2017 = pd.read_csv('./new_data/2017_US_exchange.csv', index_col=0)
df_2018 = pd.read_csv('./new_data/2018_US_exchange.csv', index_col=0)
```

```python
print("Shape of year 2014: {}".format(df_2014.shape))
print("Shape of year 2015: {}".format(df_2015.shape))
print("Shape of year 2016: {}".format(df_2016.shape))
print("Shape of year 2017: {}".format(df_2017.shape))
print("Shape of year 2018: {}".format(df_2018.shape))
```

```python
df_2014.columns
```

```python
df_2014.rename(columns={"2015 PRICE VAR [%]": "PRICE_VAR"},inplace=True)
df_2015.rename(columns={"2016 PRICE VAR [%]": "PRICE_VAR"},inplace=True)
df_2016.rename(columns={"2017 PRICE VAR [%]": "PRICE_VAR"},inplace=True)
df_2017.rename(columns={"2018 PRICE VAR [%]": "PRICE_VAR"},inplace=True)
df_2018.rename(columns={"2019 PRICE VAR [%]": "PRICE_VAR"},inplace=True)
```

```python
# Rename the train, validation and test sets
train = pd.concat([df_2014, df_2015, df_2016])
val = df_2017
test = df_2018
print(train.shape)
```

```python
train.head(5)
```

```python
train.drop(['PRICE_VAR'], axis=1, inplace=True)
train.head(5)
```

```python
# Getting the columns with'Growth'
df_growth = train.filter(regex='rowth')
df_growth.head(5)
```

```python
# Getting the columns with 'Profit'
df_profit = train.filter(regex='ofit')
df_profit.head(5)
```

```python
# Getting the columns with 'Expense'
df_expense = train.filter(regex='xpense')
df_expense.head(5)
```

```python
# Getting the columns with 'Ratio'
df_ratio = train.filter(regex='Ratio|ratio')
df_ratio.head(5)
```

```python
# Getting the columns with 'turnover'
df_turnover = train.filter(regex='Turnover|turnover')
df_turnover.head(5)
```

```python
train_features = train.iloc[:, :-1]
train_features
```

```python
train_label = train.iloc[:, -1]
train_label
```

```python
val_features = val.iloc[:, :-1]
val_label = val.iloc[:, -1]
test_features = test.iloc[:, :-1]
test_label = test.iloc[:, -1]
```

```python
# Start with only a few features: 10
features = ['Revenue', 'Net Profit Margin', 'grossProfitMargin', 'debtEquityRatio', 'Total current assets', 'Debt to Assets', 'Tangible Asset Value', 'Total current liabilities', 'returnOnEquity', 'Operating Income Growth']
X_train_10 = train_features[features]
X_train_10.head(5)
```

```python
X_train_10.describe()
```

```python
X_train_10.replace([np.inf, -np.inf], np.nan)
X_train_10.fillna(0, inplace=True)
```

```python
X_train_10.isnull().sum().sum()
np.where(np.isnan(X_train_10))
```

```python
X_val_10 = val_features[features]
X_test_10 = test_features[features]
X_val_10.replace([np.inf, -np.inf], np.nan)
X_val_10.fillna(0, inplace=True)
X_test_10.replace([np.inf, -np.inf], np.nan)
X_test_10.fillna(0, inplace=True)
```

```python
# Import classifiers
# Start with Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```python
# Start with the default parameters. Will change max_depth later
rf_model = RandomForestClassifier(random_state=0)
rf_model.fit(X_train_10, train_label)

# Prediction
train_pred = rf_model.predict(X_train_10)
train_acc = accuracy_score(train_label, train_pred)
train_f1 = f1_score(train_label, train_pred)
print("Train Accuracy: {:.4f}".format(train_acc))
print("Train F1 Score: {:.4f}".format(train_f1))

val_pred = rf_model.predict(X_val_10)
val_acc = accuracy_score(val_label, val_pred)
val_f1 = f1_score(val_label, val_pred)
print("Val Accuracy: {:.4f}".format(val_acc))
print("Val F1 Score: {:.4f}".format(val_f1))

test_pred = rf_model.predict(X_test_10)
test_acc = accuracy_score(test_label, test_pred)
test_f1 = f1_score(test_label, test_pred)
print("Test Accuracy: {:.4f}".format(test_acc))
print("Test F1 Score: {:.4f}".format(test_f1))

print("Test Confusion Matrix: ")
tn, fp, fn, tp = confusion_matrix(test_label, test_pred).ravel()
print(confusion_matrix(test_label, test_pred))
print("Test False Discovery Rate: {:.4f}".format(fp/(fp+tn)))
# Pretty bad False Discovery Rate. Skewed dataset
```

```python
# Start with the default parameters. Will change max_depth later
rf_model = RandomForestClassifier(max_depth=4, random_state=0)
rf_model.fit(X_train_10, train_label)

# Prediction
train_pred = rf_model.predict(X_train_10)
train_acc = accuracy_score(train_label, train_pred)
train_f1 = f1_score(train_label, train_pred)
print("Train Accuracy: {:.4f}".format(train_acc))
print("Train F1 Score: {:.4f}".format(train_f1))

val_pred = rf_model.predict(X_val_10)
val_acc = accuracy_score(val_label, val_pred)
val_f1 = f1_score(val_label, val_pred)
print("Val Accuracy: {:.4f}".format(val_acc))
print("Val F1 Score: {:.4f}".format(val_f1))

test_pred = rf_model.predict(X_test_10)
test_acc = accuracy_score(test_label, test_pred)
test_f1 = f1_score(test_label, test_pred)
print("Test Accuracy: {:.4f}".format(test_acc))
print("Test F1 Score: {:.4f}".format(test_f1))
```
