```python
import pandas as pd
import numpy as np
df_2014 = pd.read_csv('./new_data/2014_US_exchange.csv')
df_2015 = pd.read_csv('./new_data/2015_US_exchange.csv')
df_2016 = pd.read_csv('./new_data/2016_US_exchange.csv')
df_2017 = pd.read_csv('./new_data/2017_US_exchange.csv')
df_2018 = pd.read_csv('./new_data/2018_US_exchange.csv')
```

```python
df_2014.head()
```

```python
y = df_2014['Class']
symbol = df_2014['Symbol']
sector = df_2014['Sector']
df_2014.drop(['Class', 'Symbol', 'Sector'], axis=1, inplace=True)
df_2014 = df_2014.fillna(0)
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(df_2014.values, 
                                                    y,test_size=0.3, 
                                                    random_state=1)

knn = KNeighborsClassifier(n_neighbors=43)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

knn_pred = knn.predict(X_test_scaled)
print (f" Accuracy for k-nn is {accuracy_score(y_test, knn_pred)}")
```
