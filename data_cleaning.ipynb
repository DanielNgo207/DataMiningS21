```python
df_2014 = pd.read_csv('./original_data/2014_Financial_Data.csv')
df_2015 = pd.read_csv('./original_data/2015_Financial_Data.csv')
df_2016 = pd.read_csv('./original_data/2016_Financial_Data.csv')
df_2017 = pd.read_csv('./original_data/2017_Financial_Data.csv')
df_2018 = pd.read_csv('./original_data/2018_Financial_Data.csv')

df_list = [df_2014,df_2015,df_2016,df_2017,df_2018]
df_stable = []
for df in df_list:
    df.rename({df.columns[0] : "Symbol"}, axis =1, inplace = True)
    df_stable = df_stable + [pd.DataFrame()]
Sentinel = False
i=0
for df_year in df_list:
    for symbol in df_year["Symbol"].unique():  #no duplicate symbols
        if df_NYSE[df_NYSE["ACT Symbol"] == symbol].size: #will be 0 if symbol is not in the NYSE
            #print(symbol)
            Sentinel = True
        elif df_NASDAQ[df_NASDAQ["Symbol"] == symbol].size:
            #print(symbol)
            Sentinel = True

        if Sentinel:
            if df_stable[i].empty:
                df_stable[i] = df_year[df_year["Symbol"] == symbol]
            else:
                df_stable[i] = df_stable[i].append(df_year[df_year["Symbol"] == symbol])
        Sentinel = False
    i+=1
```

```python
i=0
for df in df_stable:
    df.to_csv(f'./new_data/{2014+i}_US_exchange.csv')
    i+=1
```
